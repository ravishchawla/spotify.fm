```python
import json
import spotipy
import spotipy.util as util
import pylast
```

```python
class Finals:
    DEFAULT_FILE_URL = 'client_details.json';
    SCOPE = 'user-library-read user-follow-read';
    REDIRECT_URL = 'https://github.com/ravishchawla/spotify.fm';
    
    CLIENT_ID = 'client_id';
    CLIENT_SECRET = 'client_secret';
    USER_NAME = 'user_name';
    PASS_WORD = 'password';
    
    ARTIST_ID = 'id';
    ARTIST_NAME = 'name';
    ARTIST_HREF = 'href';
    ARTIST_IMGS = 'images';
    ARTIST_IMG_URL = ['images', 0, 'url'];
    ARTIST_URL_BY_SONG = ['external_urls', 'spotify'];
    SONG_ID = 'id';
    SONG_NAME = 'name';
    SONG_ARTIST = 'artists';
    SONG_HREF = 'href';
    
```

```python
class spotifyfm(object):
    
    def __init__(self):
        pass;
    
    #Returns auth details
    def load_auth_details(self, file_url=Finals.DEFAULT_FILE_URL):
        try:
            file = open(file_url, 'r');
            self.auth_details = json.load(file);
        except IOError:
            print('File Load error');

    #authenticate Spotify app
    def authenticate_spotify_app(self):
        self.load_auth_details();
        spotify_details = self.auth_details['spotify'];
        token = util.prompt_for_user_token(username=spotify_details[Finals.USER_NAME], \
                                           client_id=spotify_details[Finals.CLIENT_ID], \
                                           client_secret=spotify_details[Finals.CLIENT_SECRET], \
                                           scope=Finals.SCOPE, redirect_uri=Finals.REDIRECT_URL);
        self.spotapp = spotipy.Spotify(auth=token);


        return;

    #authenticate LastFM app
    def authenticate_fm_app(self):
        self.load_auth_details();
        fm_details = self.auth_details['lastfm'];
        self.fmapp = pylast.LastFMNetwork(api_key=fm_details[Finals.CLIENT_ID], \
                                       api_secret=fm_details[Finals.CLIENT_SECRET], \
                                       username=fm_details[Finals.USER_NAME], \
                                       password_hash=pylast.md5(fm_details[Finals.PASS_WORD]));
        return;
    
        #Return a List of all artists on Last FM 
    def get_artists_on_fm():
        return;

    #Return a List of all artists followed on Spotify
    def get_artists_on_spotify(self):
        try:
            self.spotapp;
        except:
            self.spotapp = self.authenticate_spotify_app();
        
        collection = list();
        last_id = None;
        num_sets = 0;
        
        artists = self.spotapp.current_user_followed_artists(after=last_id);
        arts = artists['artists']['items'];
        
        while len(arts) > 0:
            for art in arts:
                datapack = {Finals.ARTIST_ID : art[Finals.ARTIST_ID], 
                   Finals.ARTIST_HREF : art[Finals.ARTIST_HREF],
                    Finals.ARTIST_IMG_URL[0] : art[Finals.ARTIST_IMG_URL[0]][Finals.ARTIST_IMG_URL[1]][Finals.ARTIST_IMG_URL[2]],
                    Finals.ARTIST_NAME : art[Finals.ARTIST_NAME]
                    }

                collection.append(datapack);
                last_id = art[Finals.ARTIST_ID];
            
            num_sets = num_sets + 1;
            
            artists = self.spotapp.current_user_followed_artists(after=last_id);
            arts = artists['artists']['items'];
        
        print('Total followed Artists : ' + str(len(collection)));
        
        return collection;
    
    #Return a List of all artists followed on Spotify
    def get_saved_artists_on_spotify(self):
        try:
            self.spotapp;
        except:
            self.spotapp = self.authenticate_spotify_app();
        
        collection = list();
        last_offset = 0;
        num_sets = 0;
        
        tracks = self.spotapp.current_user_saved_tracks(offset=last_offset);
        songs = tracks['items'];
        
        artist_ids = list();
        
        while len(songs) > 0:
            
            for song in songs:
                for art in song['track']['artists']:
                    if art['id'] not in artist_ids:
                        artist_ids.append(art['id']);
                        datapack = {Finals.ARTIST_NAME : art[Finals.ARTIST_NAME],
                               Finals.ARTIST_HREF : art[Finals.ARTIST_HREF],
                               Finals.ARTIST_ID : art[Finals.ARTIST_ID],
                                Finals.ARTIST_URL_BY_SONG[0] : art[Finals.ARTIST_URL_BY_SONG[0]][Finals.ARTIST_URL_BY_SONG[1]],
                               }
                    
                        collection.append(datapack);
            
            last_offset = last_offset + len(songs);
            num_sets = num_sets + 1;
            
            tracks = self.spotapp.current_user_saved_tracks(offset=last_offset);
            songs = tracks['items'];
        
        print('Total followed Artists : ' + str(len(collection)));
        
        return collection;
    
    def get_saved_tracks_on_spotify(self):
        try:
            self.spotapp;
        except:
            self.spotapp = self.authenticate_spotify_app();
        
        collection = list();
        last_offset = 0;
        num_sets = 0;
        
        tracks = self.spotapp.current_user_saved_tracks(offset=last_offset);
        songs = tracks['items'];
        
        while len(songs) > 0:
            for song in songs:
                song_data = song['track'];
                datapack = {Finals.SONG_ID : song_data[Finals.SONG_ID],
                           Finals.SONG_NAME : song_data[Finals.SONG_NAME],
                           Finals.SONG_HREF : song_data[Finals.SONG_HREF],
                            Finals.SONG_ARTIST : song_data[Finals.SONG_ARTIST][0][Finals.ARTIST_NAME]
                           }

                collection.append(datapack);
        
            last_offset = last_offset + len(songs);
            num_sets = num_sets + 1;
            
            tracks = self.spotapp.current_user_saved_tracks(offset=last_offset);
            songs = tracks['items'];
        
        print('Total Saved songs : ' + str(len(collection)));
        
        return collection;
        
    #Will favorite an artist on Last FM
    def fav_artist_on_fm(artist_name):
        return;

    #Will follow an artist on Spotify
    def follow_artist_on_spotify(artist_name):
        return;

    #Will favorite all artists followed on Spotify on Last FM
    def fav_all_on_fm():
        return;

    #Will follow all artists faved on Last Fm on Spotify
    def follow_all_on_spotify():
        return;

    #Will sync both accounts. If any artist is followed on one, it will be liked on the other
    def sync_all_artists():
        return;
```
